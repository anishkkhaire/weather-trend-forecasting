```python
import pandas as pd

df['last_updated'] = pd.to_datetime(df['last_updated'], errors='coerce')

df = df.dropna(subset=['last_updated', 'temperature_celsius'])

df.head()
```

```python
city = df['location_name'].value_counts().idxmax()
city
```

```python
city_df = df[df['location_name'] == city].copy()
city_df = city_df.sort_values('last_updated')
city_df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(city_df['last_updated'], city_df['temperature_celsius'])
plt.title(f"Temperature Trend - {city}")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
ts = (
    city_df
    .set_index('last_updated')['temperature_celsius']
    .resample('D')
    .mean()
    .interpolate()
)

ts.head(), ts.tail()
```

```python
train = ts.iloc[:-30]
test = ts.iloc[-30:]
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(train, order=(1,1,1), seasonal_order=(1,1,1,7))
model_fit = model.fit(disp=False)

forecast = model_fit.forecast(steps=30)
forecast
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(train.index, train, label="Train")
plt.plot(test.index, test, label="Actual")
plt.plot(test.index, forecast, label="Forecast")
plt.legend()
plt.title(f"30-Day Temperature Forecast - {city}")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(test, forecast)
rmse = np.sqrt(mean_squared_error(test, forecast))

mae, rmse
```

```python
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt

clf = IsolationForest(contamination=0.02, random_state=42)
flags = clf.fit_predict(ts.values.reshape(-1, 1))

anomalies = ts[flags == -1]

plt.figure()
plt.plot(ts, label="Temperature")
plt.scatter(anomalies.index, anomalies, color='red', label="Anomaly")
plt.legend()
plt.title("Temperature Anomaly Detection")
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomp = seasonal_decompose(ts, model='additive', period=7)
decomp.plot()
plt.show()
```

```python
import seaborn as sns

aq_df = city_df[['temperature_celsius', 'air_quality_PM2.5']].dropna()

corr = aq_df.corr()
corr
```

```python
plt.figure()
sns.scatterplot(
    x='air_quality_PM2.5',
    y='temperature_celsius',
    data=aq_df
)
plt.title("Temperature vs PM2.5")
plt.xlabel("PM2.5")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

X = city_df[['humidity', 'wind_kph', 'pressure_mb', 'air_quality_PM2.5']].dropna()
y = city_df.loc[X.index, 'temperature_celsius']

model = RandomForestRegressor(random_state=42)
model.fit(X, y)

importances = pd.Series(model.feature_importances_, index=X.columns)
importances.sort_values().plot(kind='barh', title='Feature Importance')
plt.show()
```
